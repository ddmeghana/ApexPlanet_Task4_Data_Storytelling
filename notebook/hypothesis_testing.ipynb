### Hypothesis Testing Notebook

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("../data/cleaned_superstore.csv")
df.head()
```

### Using the following hypothesis

#### Business Hypothesis

##### ->High discounts reduce profitability.

#### Null Hypothesis H₀

##### ->There is no significant difference in profit between low-discount and high-discount orders.

#### Alternative Hypothesis H₁

##### ->High-discount orders have significantly lower profit than low-discount orders.

```python
# Creating discount groups
low_discount = df[df['Discount'] <= 0.2]['Profit']
high_discount = df[df['Discount'] > 0.2]['Profit']

print("Low discount count:", len(low_discount))
print("High discount count:", len(high_discount))
```

```python
#Perform T test
t_stat, p_value = stats.ttest_ind(low_discount, high_discount, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
#Interpret result
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is a statistically significant difference in profit between low-discount and high-discount orders.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is no statistically significant difference in profit between the two groups.")
```

```python
#Compare Average Profit
print("Average Profit - Low Discount:", low_discount.mean())
print("Average Profit - High Discount:", high_discount.mean())
```
